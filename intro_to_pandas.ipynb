## **GroupBy**
gourps rows based off a column and performs a function on the data.

we can group a set of columns depends on values of one of them and perform a function on all data by ```df.groupby('column name to aggregate the data according to its value').function()``` like ```credit_data.groupby('default').mean()``` to get the mean of all columns according to ```1,0``` the values within column ```default```.

we can also do aggregate function on specific columns of dataframe like ``` credit_data.groupby('default')[['EDUCATION', 'MARRIAGE']].max()```

````python
credit_data.groupby('default').mean()
````

````python
# find max value of EDUCATION AND MARRIAGE by default
credit_data.groupby('default')[['EDUCATION', 'MARRIAGE']].max()
````

````python
credit_data.groupby('default')[['EDUCATION', 'MARRIAGE']].sum()
````

##### **Aggregation function**
Aggregate using one or more operations over the specified axis. ```df.agg(func=None, axis=0, *args, **kwargs)```
like ```credit_data.agg('mean')``` it give the mean functionsof all columns. we can all add more then one function ```credit_data.agg(['mean','sum'])``` you even can add a fuction you wrote it yourself.

as before we could combine between ```groupby``` and ```.agg()```.

````python
credit_data.agg(['count', 'mean'])
````

````python
def square_roots(p):
  return np.mean(np.sqrt(p))
credit_data.groupby('default')['LIMIT_BAL'].agg(['count', 'mean', square_roots])
````

## **Merging Data**


````python
# construct 4 tables of data
# employee first name
data1 ={
    'employee_id':[1,2,3,4,5,6,7],
    'first_name' :['Nathan','Jane','Judy','Kimmel','Brian','Liz','Trishton']
}
# job
data2 ={
    'employee_id':[6,5,4,3,2,1],
    'job_title':['Engineer','Accountant','Developer','Developer','Director','Manager']
}
# employee last name
data3 ={
    'employee_id' : [4,5,3,2,6,1,8],
    'last_name':['Abebe','Trent','Park','Lim','Wood','Muller','Tan']
}
# hire year
data4 = {
    'Employee_id':[3,4,1,2,5,6],
    'hire_year' :[2012,2022,2015,2000,2020,2010]
}
# convert all tables into dataframe
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
df4 = pd.DataFrame(data4)
````

Merging tables by ```pd.merge(left= left dataframe, right= right dataframe, on = 'name of common column between them')```

````python
# merge data on employee_id like SQL  
# combine first and last name
pd.merge(left = df1, right= df3, on ="employee_id")
````

Merging data has more than method according to which part you want to include within your data 
1. if you want all the left table and the intersection between the two tables you will use left merge
2. if you want all the right table and the intersection between the two tables you will use right merge
3. if you want only the intersection between the two tables you will use inner merge
4. if you want the  whole left and right table you will use outer merge
1_9eH1_7VbTZPZd9jBiGIyNA.png

we noticed that there is missing value from the first table so, we will use the left merge to get all the data within the first table first. ```pd.merge(left= left dataframe, right= right dataframe, on = 'name of common column between them', how =' how you want to merge them', indicator = True or false)``` we use indicator to see the sours=ce of merging data.

````python
pd.merge(left = df1, right = df3, how ='left', on= 'employee_id', indicator = True)
````

````python
pd.merge(left = df1, right = df3, how ='right', on= 'employee_id', indicator = True)
````

````python
pd.merge(left = df1, right = df3, how ='inner', on= 'employee_id', indicator = True)
````

````python
pd.merge(left = df1, right = df3, how ='outer', on= 'employee_id', indicator = True)
````

So, we use the prevoius way to merge two table but if we have more then two. It is the same idea you merge every two table together and but them into a temporary dataframe and merge all together at the end.

````python
# temp df between df1 anf df3 
df5 = pd.merge(left = df1, right = df3, how ='outer', on= 'employee_id')
# the temp df between the df5 and df2
df6 = pd.merge(left = df5, right = df2, how ='outer', on= 'employee_id')
````

## **Concatnation**
It differs from **Merging** two tables shouldn't have any common columns, we just but two table together. ```pd.conact([first dataframe, second dataframe])```

**Concat on the axis = 0 (rows)**
put the tables under each other

````python
pd.concat([df1,df2])
````

**Concat on the axis = 1 (columns)**
 put the tables side by side.

````python
pd.concat([df1,df2], axis = 1)
````

concatnation is useful to pusha new rows to an exsiting dataframe.

````python
# example of adding one row to a dataframe
df_new = pd.DataFrame(data = [[9, 'Holly']],
                      columns = ['employee_id', 'first_name'])
pd.concat([df1, df_new], ignore_index= True) 
````

Adding more than one row to a dataframe

````python
ids = [12,19,20,14,16]
persons = ['Timothy', 'Mathias', 'spencer', 'Alice', 'joy']
for id, person in zip(ids, persons):
  temp_df = pd.DataFrame(data = [[id, person]],
                         columns = ['employee_id', 'first_name'])
  df1 = pd.concat([df1,temp_df], ignore_index =True)

````
