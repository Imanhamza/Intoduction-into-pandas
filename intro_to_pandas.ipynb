###### **Filtering**
To use filtering we can think the filter is the new column we want to extract from the data so we do it like that ``` df[filter condition]```:
if we want to filter the filter the results where the region B within ```wholesale_data``` we can do it like that ``` wholesale_data[wholesale_data['Region'] == 'B'``` 

````python
# filter all data where Frozen > 5000
wholesale_data[wholesale_data['Frozen'] > 5000]
````

fro apply multiple filter we but every condition within ```()``` and use logical operators.

````python
# apply multiple filters 
wholesale_data[(wholesale_data['Frozen'] > 5000) & (wholesale_data['Milk'] < 600)]
````

````python
# use string within filter 
wholesale_data[wholesale_data['Channel'] == 'A']
````

````python
# use multiple filter and multiple strings
wholesale_data[(wholesale_data['Channel'].isin(['A', 'B'])) & (wholesale_data['Milk'] < 600)]
````

````python
# findout datatype of each column
wholesale_data.info()
````

````python
# change datatype of a column
wholesale_data['Grocery'] = wholesale_data['Grocery'].astype('str')
````

````python
wholesale_data.info()
````

Finding a value by knowing the its location by using 
1. **``` df.loc[]```**:
if you want all the value within a row use ```df.loc[index(row number)]``` butif you want a specific value use ```df.loc[index(row number), 'column name']``` or 
2. **```df.iloc[]```**:
the amin differnece between ```loc[]``` and ```iloc[]``` that ```iloc[]``` take a numeric values only like -> ```df.iloc[index(row number), index(column number)]```

````python
# find the third value of the 'Milk' column by using loc[]
wholesale_data.loc[2, 'Milk']
````

````python
# all value at index 5
wholesale_data.loc[5]
````

````python
# find a specific number by using iloc[]
wholesale_data.iloc[2,3]
````

we also could change the value by knowing its location ``` df.loc[index(row number), 'column name'] = new value``` 

````python
wholesale_data.shape
````

````python
for index in range(440):
  wholesale_data.loc[index, 'Milk'] = index
````

##### **String Functions**

String has some functions to deal with them in the data like ```contains('str')``` to find all the word that contain the specific word you want to search.

````python
# extrect all the countery from 'Height of Male and Female by Country 2022.csv'
# that contains 'land' in its name
df[df['Country Name'].str.contains('land')] # .str to convert the sereis into string 
````

Strings has another function to let you convert value to another value called ```.replace('original str', 'new str')```  

````python
# convert word 'land with another word
df['Country Name'].str.replace('land', 'ice')
````

````python
df
````

Dataframe hasn't change even after the run, cause we didn't update the value of the column.

````python
# update the dataframe with the replaced word
df['Country Name'] = df['Country Name'].str.replace('land', 'ice')
````

##### **Sorting**
Operator that allows you to sort your dataset according to your repective columns.

````python
credit_data = pd.read_csv('/content/credit.csv')
````

````python
credit_data.head()
````

to sort the values we use ```.sort_values(by = ['column name']) ``` this sorts values in ascending order by default if we want to reverse the order we use ```.sort_values(by = ['column name'], ascending = False)```
also we can use multiple columns to sort the data ``` .sort_values(by = ['column name 1', 'column name 2', 'column name 3']) ```

````python
# sorting the credit_data by 'Education' column
credit_data.sort_values(by = ['EDUCATION'])
````

````python
# sort by descending order
credit_data.sort_values(by= ['EDUCATION'], ascending = False)
````

````python
# sort with multiple columns
credit_data.sort_values(by = ['EDUCATION', 'AGE', 'MARRIAGE'])
````
